```python
import numpy as np
np.random.seed(0)
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
X, y = datasets.make_moons(n_samples=1000, noise=0.1, random_state=0)
y.shape, X.shape
```

```python
colors = ['steelblue' if label == 1 else 'darkred' for label in y]
plt.scatter(X[:, 0], X[:, 1], c=colors)
```

```python
model =  Sequential()
model.add(Dense(5,activation='relu', input_shape=(2,)))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
sgd = SGD(learning_rate=0.01)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])
```

```python
history = model.fit(X[:500],y[:500], verbose=1, epochs=1000, shuffle=True)
```

```python
plt.plot(history.history['loss'])
```
